```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# This list will store all the books from every page
book_data = []

# Loop through all 50 pages
for page in range(1, 51):

    # Create the URL for the current page
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"

    # Send a request to the website
    response = requests.get(url)

    # Convert the HTML into a BeautifulSoup object
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all book containers on the current page
    books = soup.find_all("article", class_="product_pod")

    # Loop through every book on the page
    for book in books:

        # Extract the title
        title = book.find("h3").find("a")["title"]

        # Extract the price
        price = book.find("p", class_="price_color").text

        # Extract availability
        availability = book.find(
            "p", class_="instock availability"
        ).text.strip()

        # Extract the rating
        rating = book.find("p", class_="star-rating")["class"][1]

        # Store the extracted information in a dictionary
        book_info = {
            "Title": title,
            "Price": price,
            "Availability": availability,
            "Rating": rating
        }

        # Add the dictionary to the list
        book_data.append(book_info)

# Convert the list into a DataFrame
df = pd.DataFrame(book_data)

# Display the first five rows
df.head()



```

```python
# Display the total number of books scraped
print(f"Total books scraped: {len(df)}")

# Save the data as a CSV file
df.to_csv("books.csv", index=False)

print("Scraping completed successfully!")
```
