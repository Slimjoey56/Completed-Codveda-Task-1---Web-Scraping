```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/cleaned_books.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Price"], bins=20)
plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df["Price"])
plt.title("Book Price Boxplot")
plt.ylabel("Price (£)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Rating", data=df)
plt.title("Distribution of Book Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Books")
plt.show()
```

```python
correlation = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(correlation, annot=True, cmap="Blues")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Price"], df["Rating"])
plt.title("Price vs Rating")
plt.xlabel("Price (£)")
plt.ylabel("Rating")
plt.show()
```

## Conclusion

- The dataset contains 1,000 books.
- Most books are priced between £10 and £40.
- Ratings are distributed across all five categories.
- A few books have unusually high prices, identified as outliers in the boxplot.
- The correlation between Price and Rating is weak, indicating that more expensive books do not necessarily receive higher ratings.
